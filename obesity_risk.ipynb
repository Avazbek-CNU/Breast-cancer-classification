```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('obesity_level.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
from sklearn.preprocessing import  LabelEncoder
mk = LabelEncoder()
df['Gender'] = mk.fit_transform(df['Gender'])
df['CALC'] = mk.fit_transform(df['CALC'])
df['MTRANS'] = mk.fit_transform(df['MTRANS'])
df['0be1dad'] = mk.fit_transform(df['0be1dad'])
df['CAEC'] = mk.fit_transform(df['CAEC'])
```

```python
df
```

```python
df.drop(columns=['id'], inplace=True)
```

df.iloc[:, :-1] selects all rows (:) and all columns except the last one (:-1) to create the feature variables.

df.iloc[:, -1] selects all rows (:) and only the last column (-1) to create the target variable.
Now, X contains your feature variables, and y contains your target variable.

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

#The model is correct in its predictions approximately 85.5% of the time.
